```python
import pandas as pd
import os

# Paths to your four CSV files
file1 = r"C:\Users\32011\Documents\ISB Work\Loans\category_1_yearly_avg.csv"
file2 = r"C:\Users\32011\Documents\ISB Work\Loans\category_2_yearly_avg.csv"
file3 = r"C:\Users\32011\Documents\ISB Work\Loans\category_3_yearly_avg.csv"
file4 = r"C:\Users\32011\Documents\ISB Work\Loans\category_4_yearly_avg.csv"

files = [file1, file2, file3, file4]
dfs = [pd.read_csv(f) for f in files]

# Assume first column is the common key
key_col = dfs[0].columns[0]

# Function to merge based on a specific column index and keep original names
def merge_on_column(col_index):
    selected_dfs = []
    for df in dfs:
        col_name = df.columns[col_index]
        selected = df[[key_col, col_name]]  # keep original name
        selected_dfs.append(selected)
    
    merged = selected_dfs[0]
    for other in selected_dfs[1:]:
        merged = merged.merge(other, on=key_col, how="outer")
    
    return merged

# Folder to save
output_dir = os.path.dirname(file1)

# Create 3 merged CSVs (3rd, 4th, and 5th columns  index 2,3,4)
for idx in [2, 3, 4, 5, 6]:
    merged_df = merge_on_column(idx)
    output_path = os.path.join(output_dir, f"merged_column_{idx}.csv")
    merged_df.to_csv(output_path, index=False)
    print(f" Saved merged file for column {idx}: {output_path}")
```
